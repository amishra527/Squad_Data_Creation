```python
# import pandas as pd

df = pd.read_json('new21.json', orient='records')

def check_answer_start(row):
    answer_starts = row['answers']['answer_start']
    for answer_start in answer_starts:
        if not isinstance(answer_start, int):
            print(f'Invalid value for answer_start: {answer_start}')

df.apply(check_answer_start, axis=1)
```
