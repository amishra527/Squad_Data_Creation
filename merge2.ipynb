```python
# puts the required formant
import json
import csv
import ast

with open('new19.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

with open('new19.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'title', 'context', 'question', 'answers'])
    writer.writeheader()
    for item in data:
        answers = ast.literal_eval(item['answers'])
        row = {
            'id': item['id'],
            'title': item['title'],
            'context': item['context'],
            'question': item['question'],
            'answers': {'text': [ans['text'] for ans in answers], 'answer_start': [ans['answer_start'] for ans in answers]}
        }
        writer.writerow(row)
```
